```python
import os
import pandas as pd
import re

# ---------------------------------------
# FOLDER WHERE INDIVIDUAL CSV FILES ARE SAVED
# ---------------------------------------
input_folder = r"C:/Users/Asus/Pictures/Lenskart"   # CHANGE THIS
output_file = r"C:/Users/Asus/Pictures/Lenskart/final_products.csv"

# ---------------------------------------
# FUNCTION TO CLEAN CATEGORY NAME
# ---------------------------------------
def clean_category(filename):
    name = os.path.splitext(filename)[0]  # remove .csv

    name = name.replace("lenskart_", "")        # remove prefix
    name = name.replace("all-", "")             # remove unnecessary 'all'
    name = name.replace("-", " ")               # replace - with space

    # Remove duplicate words (if any)
    words = name.split()
    cleaned_words = []
    for w in words:
        if w not in cleaned_words:
            cleaned_words.append(w)

    clean_name = " ".join(cleaned_words).title()   # Title case: "Kids Eyeglasses"
    return clean_name


# ---------------------------------------
# READ AND MERGE CSV FILES
# ---------------------------------------
all_dataframes = []

for filename in os.listdir(input_folder):
    if filename.endswith(".csv"):
        file_path = os.path.join(input_folder, filename)

        category_name = clean_category(filename)

        print(f"Processing: {filename}  ‚Üí  Category: {category_name}")

        df = pd.read_csv(file_path)
        df["Category"] = category_name

        all_dataframes.append(df)

# MERGE ALL DATA
final_df = pd.concat(all_dataframes, ignore_index=True)
final_df.drop_duplicates(inplace=True)
final_df.to_csv(output_file, index=False)

print("\n‚úÖ Final CSV created successfully!")
print(f"üìÅ Saved at: {output_file}")
```
